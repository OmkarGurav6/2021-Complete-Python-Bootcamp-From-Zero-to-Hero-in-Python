```python
# one_star_titles = []

# for n in range(1,51):

#     scrape_url = base_url.format(n)
#     res = requests.get(scrape_url)
    
#     soup = bs4.BeautifulSoup(res.text,"lxml")
#     books = soup.select(".product_pod")
    
#     for book in books:
#         if len(book.select('.star-rating.One')) != 0:
#             one_star_titles.append(book.select('a')[1]['title'])
star_titles = []

def title_star(stars):
    
        if len(star_titles) != 0:
            star_titles.clear()
         
        for n in range(1,51):

            scrape_url = base_url.format(n)
            res = requests.get(scrape_url)

            soup = bs4.BeautifulSoup(res.text,"lxml")
            books = soup.select(".product_pod")

            for book in books:
               
                if stars == 'One':
                    if len(book.select('.star-rating.One')) != 0:
                        star_titles.append(book.select('a')[1]['title'])

                elif stars == 'Two':
                    if len(book.select('.star-rating.Two')) != 0:
                        star_titles.append(book.select('a')[1]['title'])

                elif stars == 'Three':
                    if len(book.select('.star-rating.Three')) != 0:
                        star_titles.append(book.select('a')[1]['title'])
                        
                elif stars == 'Four':
                    if len(book.select('.star-rating.Four')) != 0:
                        star_titles.append(book.select('a')[1]['title'])
                        
                else:
                    if len(book.select('.star-rating.Five')) != 0:
                        star_titles.append(book.select('a')[1]['title'])
```

```python
title_star(stars = 'One')
```
