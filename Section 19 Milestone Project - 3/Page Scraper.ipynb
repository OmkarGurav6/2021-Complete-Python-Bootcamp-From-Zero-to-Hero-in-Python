```python
# Create an application which connects to a site and pulls out all links, or images, and saves them to a list.

import requests, bs4
import validators
```

```python

def check_Internet_Requests(url='http://www.google.com/', timeout=3):
    
    try:
        
        r = requests.get(url, timeout=timeout)
        print("Connected to the Internet")
        return True
    
    except requests.ConnectionError as ex:
        
        print("No internet connection..")
        print(ex)
        return False

    
if check_Internet_Requests():   

        valid = True


        while valid:

                url = input("Enter url of site : ")


                if validators.url(url):

                    valid = False

                    res = requests.get(url)


                    soup = bs4.BeautifulSoup(res.text,'lxml')

                    link_list = soup.find_all('a')

                    images_list = soup.find_all('img')

                    links = []
                    images = []

                    # Appending links
                    for l in link_list:

                        links.append(l.get('href'))

                    # printing all links
                    print("List of all links : \n")
                    for l in links:
                        print(l)

                    # Appending images
                    for i in images_list:

                        images.append(i.get('src'))

                    # printing all images
                    print("\nList of all images : \n")
                    for i in images:
                        print(i)


                else:

                    print("Wrong url !!! Enter again.")

```
