```python
# Prime Factorization - Have the user enter a number and find all Prime Factors (if there are any) and display them.

def is_prime(n):

    if n in [2]:
        return True
    elif n % 2 == 0 or n == 1:
        return False
        
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False 
            
    return True


def factors(n):
    
    prime = []
    
    for i in range(1,int(n/2) + 1):
        if n % i == 0:
            if is_prime(i) == True:
                prime.append(i)
                
    if is_prime(n) == True:
        prime.append(n)

    if len(prime) == 0:
        print(f"Number {n} doesn't have any prime factor.")
    else:
        print(prime)


def main():
    
    current_not_valid = True
    
    
    while current_not_valid:
        number = int(input("Enter number : "))
        
        if number <= 0:
            current_not_valid = True
            print("Wrong choice !!! Enter positive integer.")
        else:
            current_not_valid = False
            
            factors(number)

            ch = input("Do you want to enter another number ? (y/n) : ")

            if ch.lower().startswith('y'):
                current_not_valid = True
        
        
if __name__ == '__main__':
    main()
```
