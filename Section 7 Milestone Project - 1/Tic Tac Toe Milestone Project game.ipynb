```python
# Function for displaying horizontal line
def display_horizontal_line():
    print("-" + "-" + "-" + "-" + "-" + "-" + "-" + "-" + "-" + "-" + "-")

# Function for displaying board
def display_board(board):
    
    print("\n")
    print(" " + board[7] + " | " + board[8] + " | " + board[9] + " ")
    display_horizontal_line()
    print(" " + board[4] + " | " + board[5] + " | " + board[6] + " ")
    display_horizontal_line()
    print(" " + board[1] + " | " + board[2] + " | " + board[3] + " ")

```

```python
# Function for choosing marker
def player_input():
    
    marker = ""
    
    while (marker not in ["X","O"]):
           
            marker = input("Player 1, Do you want to be X or O ? : ").upper()
           
    if marker == 'X':
        return ("X","O")
    else:
        return ("O","X")
        
```

```python
# Function for assigning marker to position
def place_marker(board, marker, position):
    board[position] = marker
    
```

```python
# Function for checking if mark has won
def win_check(board, mark):
    
    return ((board[1] == mark and board[2] == mark and board[3] == mark) or #row1
            (board[4] == mark and board[5] == mark and board[6] == mark) or #row2
            (board[7] == mark and board[8] == mark and board[9] == mark) or #row3
            (board[1] == mark and board[4] == mark and board[7] == mark) or #column1
            (board[2] == mark and board[5] == mark and board[8] == mark) or #column2
            (board[3] == mark and board[6] == mark and board[9] == mark) or #column3
            (board[1] == mark and board[5] == mark and board[9] == mark) or #diagonal1
            (board[3] == mark and board[5] == mark and board[7] == mark)    #diagonal2
           )
    
```

```python
# Function for checking if position is available
def space_check(board, position):
    return board[position] == " "
```

```python
# Function for checking if board is full
def full_board_check(board):
    for pos in range(1,10):
        if space_check(board,pos):
            return False
    return True
```

```python
# Function for taking players input
def player_choice(Turn, board):
    
    position = 0
    
    while position not in range(1,10) or not space_check(board,position):
        
        position = int(input(Turn + ", Choose your next position (1-9) : "))
        
    return position
```

```python
# Function for asking if players want to play again
def replay():
    
    return (input("Do you want to play again ? Enter Yes or No : ").upper().startswith("Y"))
    
    
```

```python
# Main function
print('Welcome to Tic Tac Toe!')

while True:
    board = [" "] *10
    player_1_marker, player_2_marker = player_input()
    print("Player 1 will go first.")
   
    play_game = input("Are you ready to play? Enter Yes or No : ")
    
    if play_game.upper()[0] == "Y":
        game_on = True
    else:
        game_on = False

    Turn = "Player 1"
        
    while game_on:
        #Player 1 Turn
        if Turn == "Player 1":
            display_board(board)
            position = player_choice(Turn, board)
            place_marker(board, player_1_marker, position)
                
            if win_check(board, player_1_marker):
                display_board(board)
                print("Congrats ! Player 1 won the game !!!")
                game_on = False
            else:
                if full_board_check(board):
                    display_board(board)
                    print("The game is a draw !")
                    break
                    
            Turn = "Player 2"
        
        # Player2's turn.
        else:
            display_board(board)
            position = player_choice(Turn, board)
            place_marker(board, player_2_marker, position)
                
            if win_check(board, player_2_marker):
                display_board(board)
                print("Congrats ! Player 2 won the game !!!")
                game_on = False
            else:
                if full_board_check(board):
                    display_board(board)
                    print("The game is a draw !")
                    break
        
            Turn = "Player 1"

    if not replay():
        break
```
